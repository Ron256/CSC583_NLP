```python
def cosineSimilarity(word1, word2):
  """Function  that returns the cosine similarity value of two vectors """
  a = wordVectors[word1]
  b = wordVectors[word2]

  return np.dot(a, b) /(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
def computeDogSimilarity():
  """Function that returns a populated dictionary of different similar words and theirs cosine similarity values"""
  cosine_sim = {}
  for key, value in wordVectors.items():
    if key == 'dogs':
      pass
    elif key == 'dog':
      pass
    else:
      cs_value = cosineSimilarity('dog', key)
      word = 'dog' + '-' + key
      cosine_sim[word] = cs_value
  return cosine_sim
```

```python
def computeWhaleSimilarity():
  """Function that returns a populated dictionary of different similar words and theirs cosine similarity values"""
  cosine_sim = {}
  for key, value in wordVectors.items():
    if key == 'whales':
      pass
    elif key == 'whale':
      pass
    else:
      cs_value = cosineSimilarity('whale', key)
      word = 'whale' + '-' + key
      cosine_sim[word] = cs_value
  return cosine_sim
```

```python
def computeBeforeSimilarity():
  """Function that returns a populated dictionary of different similar words and theirs cosine similarity values"""
  cosine_sim = {}
  for key, value in wordVectors.items():
    if key == 'after': #('after', 'afterwards', 'ahead', 'before', 'when')
      pass
    elif key == 'afterwards':
      pass
    elif key == 'ahead':
      pass
    elif key == 'when':
      pass
    elif key == 'before':
      pass
    else:
      cs_value = cosineSimilarity('before', key)
      word = 'before' + '-' + key
      cosine_sim[word] = cs_value
  return cosine_sim
```

```python
def computeHoweverSimilarity():
  """Function that returns a populated dictionary of different similar words and theirs cosine similarity values"""
  cosine_sim = {}
  for key, value in wordVectors.items():
    if key == 'however':
      pass
    else:
      cs_value = cosineSimilarity('however', key)
      word = 'however' + '-' + key
      cosine_sim[word] = cs_value
  return cosine_sim
```

```python
def computeFabricateSimilarity():
  """Function that returns a populated dictionary of different similar words and theirs cosine similarity values"""
  cosine_sim = {}
  for key, value in wordVectors.items():
    if key in ['fabricated', 'fabricate', 'fabricates']:
      pass
    else:
      cs_value = cosineSimilarity('fabricate', key)
      word = 'fabricate' + '-' + key
      cosine_sim[word] = cs_value
  return cosine_sim
```

```python
def computeAnimalSimilarity():
  """Function that returns a populated dictionary of different similar words and theirs cosine similarity values"""
  cosine_sim = {}
  for key, value in wordVectors.items():
    if key in ('animal', 'animals', 'animales', \
               'animalia', 'animalistic', 'animals-like'):
      pass
    else:
      cs_value = cosineSimilarity('animal', key)
      word = 'animal' + '-' + key
      cosine_sim[word] = cs_value
  return cosine_sim
```

```python
df_cosineSimDog = pd.DataFrame()
def deriveMaximumValue(pd_f=df_cosineSimDog):
  """Function that prints the maximum value amongst all the different similar words"""
  # extract the cosine values into numpy
  cosine_value_L = pd_f['cosine_value'].values.tolist()
  cosine_value_np = np.array(cosine_value_L)
  max_value = cosine_value_np.max()
  # derive a filter
  filter = (pd_f['cosine_value'] == max_value)
  print(pd_f[filter])
```
